# DATA CLEANING
Memeriksa apakah ada data yang NULL atau yang Duplikat

```python
dataset_split = pd.read_csv("tourist-departures.csv")
dataset_split.head(10)
```

# Memeriksa apakah ada data yang tidak ber isi atau nilai NULL

```python
dataset_split.isna().sum()
```

# Melakukan data cleaning dengan data yang bernilai NULL

```python
from sklearn.impute import SimpleImputer
```

# Untuk mengubah data NULL dengan strategi Most Frequent

```python
imputer1 = SimpleImputer(strategy="most_frequent")
dataset_split["Code"] = imputer1.fit_transform(dataset_split[["Code"]])
dataset_split.isna().sum()
```

```python
dataset_split.head(20)
```

# Menggunakan sklearn.preprocessing MinMaxScaaler

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# Mengubah nilai NULL dengan strategi mean

```python
imputer2 = SimpleImputer(strategy="mean")
dataset_split["Outbound departures (tourists) per 1000 people"] = imputer2.fit_transform(dataset_split[["Outbound departures (tourists) per 1000 people"]])
dataset_split.isna().sum()
```

# Mengubah nilai NULL dengan strategi mean

```python
imputer3 = SimpleImputer(strategy="median")
dataset_split["GDP per capita, PPP (constant 2017 international $)"] = imputer3.fit_transform(dataset_split[["GDP per capita, PPP (constant 2017 international $)"]])
dataset_split.isna().sum()
```

## DATA ENCODING

```python
data_encoding = pd.read_csv("tourist-departures.csv")
```

# Menerapkan OneHotEncoder pada data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(sparse=False)
```

```python
Continent_encoder = onehotencoder.fit_transform(data_encoding[['Continent']])
```

```python
Continent = pd.DataFrame(Continent_encoder)
data_encoding = data_encoding.join(Continent)

data_encoding.head(10)
```

# Menerapkan get_dummies pada data

```python
Entity_encoder = pd.get_dummies(data_encoding[['Entity']])

Entity = pd.DataFrame(Entity_encoder)
data_encoding = data_encoding.join(Entity)

data_encoding.head(10)
```
