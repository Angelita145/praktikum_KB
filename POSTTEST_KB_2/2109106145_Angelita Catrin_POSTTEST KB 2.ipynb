```python
cancer = pd.read_csv('tiktok_google_play_reviews.csv')
record = cancer.shape[0]
atribut = cancer.columns
objek = cancer.select_dtypes(include=object).columns
angka = cancer.select_dtypes(include=np.number).columns
Null = cancer.loc[:, cancer.isna().any()].columns
```

## 5. Tampilkan jumlah nilai dari semua attribute yang bernilai object.

```python
# Membaca data dari file 'tiktok_google_play_reviews.csv'
cancer = pd.read_csv('tiktok_google_play_reviews.csv')

# Memilih kolom-kolom dengan tipe data objek (kategorikal)
objek = cancer.select_dtypes(include=['object'])

# Menghitung jumlah variasi (nilai unik) dalam setiap kolom kategorikal
jumlah_objek = objek.nunique()

# Membuat DataFrame untuk tabel variasi
tabel = pd.DataFrame({
    'Nama Atribut': jumlah_objek.index, 
    'Total Nilai': jumlah_objek.values})

# Menampilkan tabel variasi
print(tabel)
```
