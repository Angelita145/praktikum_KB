```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('tiktok_google_play_reviews.csv') #pendeklarasian untuk membaca data
```

```python
df.head()
```

## GRAFIK PIE CHART

Pie Chart ini digunakan untuk menghitung nilai yang banyak serta tidak, dan berbentuk lingkaran.
Dapat Dilihat bahwa pie chart dibawah ini menghitung skor dari data set yang didalamnya yaitu ada skor 2 yang paling sedikit serta skor 5 paling banyak di berikan oleh reviwers tiktok di Google

Kesimpulannya adalah dapat dilihat di pie chart ini, skor yang terbanyak yang diberikan oleh reviwers yang telah menonton content-content tiktok menimbulkan reaksi yang bagus dapat dilihat angka yang terbanyak yaitu 5 yang berarti content tersebut memuaskan.

```python
data = [df.score[df['score']<5].count(), df.score[df['score']>=5].count()]
label = ['2', '5']

figure, axes = plt.subplots(figsize=(8, 6))
axes.pie(data, 
         labels=label, 
         autopct='%.1f%%',
         shadow=True)

plt.title("Score 5 VS 2", size = 10)
plt.legend()
```

## BOX PLOT

Kesimpulan untuk box plot ini dapat dilihat bahwa penghitungan dari thumbs up count atau bisa dibilang penghitungan pemencetan jari dihitung dari 0 hingga 11 ribu, dapat diketahui bahwa banyak sekali perhitungan dari thumbs up count ini.

```python
# Membaca data dari file CSV
df = pd.read_csv('tiktok_google_play_reviews.csv')
plt.figure(figsize= (20,4))
sns.boxplot(x=df.thumbsUpCount, )
plt.show()
```

## GRAFIK HISTOGRAM

Kesimpulannya dapat dilihat bahwa dibawah ini penghitungan berdasarkan skor dan jam dan Tanggal.
Semakin tinggi sebuah batang maka semakin besar pula nilai skor pada tanggal dan jam tersebut.

```python
plt.figure(figsize=(10,10))
plt.hist(df['score'], bins=15, color='lavender', edgecolor='black')

plt.title('Data Histogram berdasarkan Skor dan Tanggal serta Jam')
plt.ylabel('score')
plt.xlabel('at')

plt.show()
```

```python
x= df.hist(bins=10, figsize=(8,8), color='grey')

plt.show()
```

## DIAGRAM HEAT MAP

Kesimpulannya dapat dilihat bahwa skor dan penghitungan pemencetan jari dihitung dalam metode diagram heat map.
Dapat dilihat berapa saja scor dan penghitungan lainnya.

```python
cor_matrix = df.corr()

sns.heatmap(cor_matrix, cmap='RdBu', annot=True, center=0)

plt.title('Heat Map Skor dan penghitungan lain')
plt.xlabel("Variabel")
plt.ylabel("Variabel")

plt.show()
```
